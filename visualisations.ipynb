```python
import os
import matplotlib.pyplot as plt
from pipeline.utils.common import get_root_folder
```

```python
loss = list()
file_path = os.path.join(get_root_folder(),"pipeline","logs","loss.txt")
```

```python
try:
    with open(file_path, 'r') as f:
        for line in f:
            loss.append(float(line.strip()))

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(loss, label='Loss')
    plt.title(f'Loss over {len(loss)} Iterations')
    plt.xlabel('Iteration')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.legend()
    plt.show()

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except ValueError:
    print("Error: Could not convert a line to a number. Ensure the file contains only numbers.")
```
